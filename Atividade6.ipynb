```python
from nltk.stem import WordNetLemmatizer
  
lemmatizer = WordNetLemmatizer()
```

**lendo dataframe**

```python
dataframe = pd.read_csv('/content/twitter-2013train-A.txt', sep='\t', encoding='utf8', names=['id', 'polarity', 'text'])
```

**Criando stop list e dicionario de emotes ;)**

```python
stop_list_nltk = nltk.corpus.stopwords.words('english')
stop_list_nltk.append('RT')
```

```python
emo = {}

emo[':)'] = 'happy'
emo['XD'] = 'happy'
emo[';)'] = 'happy'
emo['>_<'] = 'blushing'
emo[':('] = 'sad'
emo[';-;'] = 'crying'
emo[';_;'] = 'crying'
emo['\'-\''] = 'annoyed'
emo[':/'] = 'annoyed'
```

**Definindo função de tokenização que separa palavrs, remove emotes, stop words e lematiza**

```python
def tokeniza(sentence):
  tokens = re.split(r'[\s!$@#,.;]+',sentence)

  novel_tokens=[]
  for t in tokens:
    if t in emo.keys():
      novel_tokens.append(emo[t])
    elif t not in stop_list_nltk:
      novel_tokens.append(t)

  return [lemmatizer.lemmatize(t) for t in novel_tokens]

```

```python
sentences = dataframe['text']
sentences[0]
```

```python
final_tweets = [[]]
```

**Aplicando tokenização nas sentenças**

```python
for sentence in sentences:
  final_tweets.append(tokeniza(sentence))
```

```python
final_tweets
```

**Unindo tokenização ao dataframe novamente**

```python
dataframe["Tokenized text"]=final_tweets
```
