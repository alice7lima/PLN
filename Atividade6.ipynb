```python
import krippendorff
import numpy as np
import pandas as pd
from sklearn.metrics import cohen_kappa_score
```

```python
path = '/content/drive/MyDrive/members/T0020/pool'

annotator1 = []
for f in os.listdir(path):
  aux = path + '/' + str(f)
  annotator1.append(aux)

annotator1 = sorted(annotator1)
```

```python
path = '/content/drive/MyDrive/members/JonFrancis/pool'
annotator2 = []

for f in os.listdir(path):
  aux = path + '/' + str(f)
  annotator2.append(aux)

annotator2 = sorted(annotator2)
```

```python
a1_vals = []

for f in annotator1:
  with open(f, 'r') as json_file:
	  json_load = json.load(json_file)

  a1_vals.append(json_load['metas']['m_4']['value'])


a2_vals = []

for f in annotator2:
  with open(f, 'r') as json_file:
	  json_load = json.load(json_file) 

  a2_vals.append(json_load['metas']['m_4']['value'])

values = []
i = 1

for a,b in zip(a1_vals, a2_vals):
  d = 'd' + str(i)
  values.append([d, a, b])
  i += 1

print(values)
```

```python
df = pd.DataFrame(values, columns = ['texto','Label - Anotador 1','Label - Anotador 2'])

df
```

```python
corpus_transpose = []

for a in df:

    if a != 'texto':

        lista = np.array(df[a])
        corpus_transpose.append(list(lista.T))

corpus_transpose
```

```python
tags = {'positivo' : 2, 'neutro' : 1, 'negativo' : 0}

ct_int = []

for labels in corpus_transpose:
    t = []
    for doc in labels:
        t.append(tags[doc])
    ct_int.append(t)

ct_int
```

```python
print(cohen_kappa_score(corpus_transpose[0], corpus_transpose[1]))
```
